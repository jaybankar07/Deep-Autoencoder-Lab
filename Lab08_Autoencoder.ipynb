```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import glob
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA


np.random.seed(42)
tf.random.set_seed(42)

def load_chess_dataset(dataset_path, img_size=(128, 128)):
    """
    Load chess piece images from the directory structure
    
    Args:
        dataset_path: Path to the main dataset directory
        img_size: Tuple of (height, width) for resizing images
    
    Returns:
        images: numpy array of images
        labels: list of labels
        class_names: list of unique class names
    """
    images = []
    labels = []
    class_names = []
    
    
    chess_path = os.path.join(dataset_path, "chessman-image-dataset", "Chessman-image-dataset", "Chess")
    
    print(f"Looking for chess pieces in: {chess_path}")
    
    
    piece_types = [d for d in os.listdir(chess_path) if os.path.isdir(os.path.join(chess_path, d))]
    piece_types.sort()  
    class_names = piece_types
    
    print(f"Found piece types: {piece_types}")
    
    for i, piece_type in enumerate(piece_types):
        piece_path = os.path.join(chess_path, piece_type)
        print(f"\nLoading {piece_type} images...")
        
        
        image_extensions = ['*.jpg', '*.jpeg', '*.JPG', '*.JPEG', '*.png', '*.PNG']
        image_files = []
        for ext in image_extensions:
            image_files.extend(glob.glob(os.path.join(piece_path, ext)))
        
        print(f"Found {len(image_files)} {piece_type} images")
        
        for img_file in image_files:
            try:
                
                img = Image.open(img_file)
                
                
                if img.mode != 'RGB':
                    img = img.convert('RGB')
                
                
                img = img.resize(img_size)
                
                
                img_array = np.array(img) / 255.0
                
                images.append(img_array)
                labels.append(i)  
                
            except Exception as e:
                print(f"Error loading {img_file}: {e}")
                continue
    
    images = np.array(images)
    labels = np.array(labels)
    
    print(f"\nDataset loaded successfully!")
    print(f"Total images: {len(images)}")
    print(f"Image shape: {images.shape}")
    print(f"Labels shape: {labels.shape}")
    print(f"Number of classes: {len(class_names)}")
    
    return images, labels, class_names

def create_convolutional_autoencoder(input_shape, latent_dim=128):
    """
    Create a convolutional autoencoder model
    
    Args:
        input_shape: Shape of input images (height, width, channels)
        latent_dim: Dimension of the latent space
    
    Returns:
        autoencoder: Complete autoencoder model
        encoder: Encoder part only
        decoder: Decoder part only
    """
    
    
    encoder_input = keras.Input(shape=input_shape, name='encoder_input')
    
    x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(encoder_input)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    x = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    x = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    
    
    shape_before_flattening = keras.backend.int_shape(x)
    x = layers.Flatten()(x)
    encoded = layers.Dense(latent_dim, activation='relu', name='encoded')(x)
    
    
    encoder = keras.Model(encoder_input, encoded, name='encoder')
    
    
    decoder_input = keras.Input(shape=(latent_dim,), name='decoder_input')
    
    
    units = shape_before_flattening[1] * shape_before_flattening[2] * shape_before_flattening[3]
    x = layers.Dense(units, activation='relu')(decoder_input)
    x = layers.Reshape((shape_before_flattening[1], shape_before_flattening[2], shape_before_flattening[3]))(x)
    
    x = layers.Conv2DTranspose(256, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    x = layers.Conv2DTranspose(128, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    x = layers.Conv2DTranspose(64, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    x = layers.Conv2DTranspose(32, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    
    decoded = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same', name='decoded')(x)
    
    
    decoder = keras.Model(decoder_input, decoded, name='decoder')
    
    
    autoencoder_output = decoder(encoder(encoder_input))
    autoencoder = keras.Model(encoder_input, autoencoder_output, name='autoencoder')
    
    return autoencoder, encoder, decoder

def plot_sample_images(images, labels, class_names, num_samples=10):
    """
    Plot sample images from the dataset
    """
    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    axes = axes.ravel()
    
    indices = np.random.choice(len(images), num_samples, replace=False)
    
    for i, idx in enumerate(indices):
        axes[i].imshow(images[idx])
        axes[i].set_title(f"{class_names[labels[idx]]}")
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.show()

def plot_training_history(history):
    """
    Plot training and validation loss and metrics
    """
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    
    axes[0, 0].plot(history.history['loss'], label='Training Loss', color='blue')
    axes[0, 0].plot(history.history['val_loss'], label='Validation Loss', color='red')
    axes[0, 0].set_title('Model Loss')
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Loss')
    axes[0, 0].legend()
    axes[0, 0].grid(True)
    
    
    axes[0, 1].plot(history.history['mae'], label='Training MAE', color='blue')
    axes[0, 1].plot(history.history['val_mae'], label='Validation MAE', color='red')
    axes[0, 1].set_title('Model MAE')
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('MAE')
    axes[0, 1].legend()
    axes[0, 1].grid(True)
    
    
    if 'lr' in history.history:
        axes[1, 0].plot(history.history['lr'], color='green')
        axes[1, 0].set_title('Learning Rate')
        axes[1, 0].set_xlabel('Epoch')
        axes[1, 0].set_ylabel('Learning Rate')
        axes[1, 0].set_yscale('log')
        axes[1, 0].grid(True)
    else:
        axes[1, 0].text(0.5, 0.5, 'Learning Rate\nNot Available', 
                       ha='center', va='center', transform=axes[1, 0].transAxes)
        axes[1, 0].set_title('Learning Rate')
    
    
    loss_diff = np.array(history.history['val_loss']) - np.array(history.history['loss'])
    axes[1, 1].plot(loss_diff, color='orange')
    axes[1, 1].set_title('Validation - Training Loss')
    axes[1, 1].set_xlabel('Epoch')
    axes[1, 1].set_ylabel('Loss Difference')
    axes[1, 1].axhline(y=0, color='black', linestyle='--', alpha=0.5)
    axes[1, 1].grid(True)
    
    plt.tight_layout()
    plt.show()

def plot_reconstructions(autoencoder, X_test, class_names, labels_test, num_images=10):
    """
    Plot original vs reconstructed images
    """
    
    indices = np.random.choice(len(X_test), num_images, replace=False)
    test_samples = X_test[indices]
    
    
    reconstructed = autoencoder.predict(test_samples)
    
    
    fig, axes = plt.subplots(3, num_images, figsize=(20, 8))
    
    for i in range(num_images):
        
        axes[0, i].imshow(test_samples[i])
        axes[0, i].set_title(f'Original\n{class_names[labels_test[indices[i]]]}')
        axes[0, i].axis('off')
        
        
        axes[1, i].imshow(reconstructed[i])
        axes[1, i].set_title('Reconstructed')
        axes[1, i].axis('off')
        
        
        diff = np.abs(test_samples[i] - reconstructed[i])
        axes[2, i].imshow(diff)
        axes[2, i].set_title('Difference')
        axes[2, i].axis('off')
    
    plt.tight_layout()
    plt.show()

def plot_latent_space_visualization(encoder, X_test, labels_test, class_names):
    """
    Visualize latent space using t-SNE and PCA
    """
    
    encoded_imgs = encoder.predict(X_test)
    
    
    tsne = TSNE(n_components=2, random_state=42, perplexity=30)
    encoded_tsne = tsne.fit_transform(encoded_imgs)
    
    
    pca = PCA(n_components=2)
    encoded_pca = pca.fit_transform(encoded_imgs)
    
    fig, axes = plt.subplots(1, 2, figsize=(20, 8))
    
    
    colors = plt.cm.Set3(np.linspace(0, 1, len(class_names)))
    for i, class_name in enumerate(class_names):
        mask = labels_test == i
        axes[0].scatter(encoded_tsne[mask, 0], encoded_tsne[mask, 1], 
                       c=[colors[i]], label=class_name, alpha=0.7, s=50)
    axes[0].set_title('t-SNE Visualization of Latent Space')
    axes[0].set_xlabel('t-SNE 1')
    axes[0].set_ylabel('t-SNE 2')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    
    for i, class_name in enumerate(class_names):
        mask = labels_test == i
        axes[1].scatter(encoded_pca[mask, 0], encoded_pca[mask, 1], 
                       c=[colors[i]], label=class_name, alpha=0.7, s=50)
    axes[1].set_title('PCA Visualization of Latent Space')
    axes[1].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} variance)')
    axes[1].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} variance)')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return encoded_imgs

def plot_reconstruction_error_heatmap(autoencoder, X_test, labels_test, class_names):
    """
    Create heatmap of reconstruction errors by class
    """
    
    reconstructed = autoencoder.predict(X_test)
    mse_errors = np.mean((X_test - reconstructed) ** 2, axis=(1, 2, 3))
    
    
    error_matrix = []
    for class_idx in range(len(class_names)):
        class_mask = labels_test == class_idx
        class_errors = mse_errors[class_mask]
        error_matrix.append([
            np.mean(class_errors),
            np.std(class_errors),
            np.min(class_errors),
            np.max(class_errors),
            np.median(class_errors)
        ])
    
    error_matrix = np.array(error_matrix)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))
    
    
    sns.heatmap(error_matrix.T, 
                xticklabels=class_names, 
                yticklabels=['Mean', 'Std', 'Min', 'Max', 'Median'],
                annot=True, fmt='.4f', cmap='Reds', ax=ax1)
    ax1.set_title('Reconstruction Error Statistics by Class')
    ax1.set_xlabel('Chess Piece')
    ax1.set_ylabel('Statistic')
    
    
    error_data = []
    labels_data = []
    for class_idx in range(len(class_names)):
        class_mask = labels_test == class_idx
        class_errors = mse_errors[class_mask]
        error_data.extend(class_errors)
        labels_data.extend([class_names[class_idx]] * len(class_errors))
    
    import pandas as pd
    df = pd.DataFrame({'Error': error_data, 'Class': labels_data})
    sns.boxplot(data=df, x='Class', y='Error', ax=ax2)
    ax2.set_title('Distribution of Reconstruction Errors by Class')
    ax2.set_xlabel('Chess Piece')
    ax2.set_ylabel('MSE Error')
    ax2.tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()

def plot_feature_maps(encoder, sample_image, layer_names=None):
    """
    Visualize feature maps from different layers of the encoder
    """
    if layer_names is None:
        layer_names = [layer.name for layer in encoder.layers if 'conv2d' in layer.name]
    
    
    layer_outputs = [encoder.get_layer(name).output for name in layer_names]
    activation_model = keras.Model(inputs=encoder.input, outputs=layer_outputs)
    
    
    sample_image = np.expand_dims(sample_image, axis=0)
    activations = activation_model.predict(sample_image)
    
    
    for layer_name, activation in zip(layer_names, activations):
        n_features = activation.shape[-1]
        size = activation.shape[1]
        
        
        n_cols = min(8, n_features)
        n_rows = min(2, (n_features + n_cols - 1) // n_cols)
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 2, n_rows * 2))
        if n_rows == 1:
            axes = axes.reshape(1, -1)
        
        for i in range(min(n_rows * n_cols, n_features)):
            row, col = i // n_cols, i % n_cols
            axes[row, col].imshow(activation[0, :, :, i], cmap='viridis')
            axes[row, col].axis('off')
            axes[row, col].set_title(f'Feature {i+1}')
        
        
        for i in range(n_features, n_rows * n_cols):
            row, col = i // n_cols, i % n_cols
            axes[row, col].axis('off')
        
        plt.suptitle(f'Feature Maps - {layer_name}')
        plt.tight_layout()
        plt.show()

def comprehensive_evaluation(autoencoder, encoder, decoder, X_train, X_test, 
                           labels_train, labels_test, class_names, history=None):
    """
    Comprehensive evaluation and visualization of the autoencoder
    """
    print("="*60)
    print("COMPREHENSIVE MODEL EVALUATION")
    print("="*60)
    
    
    if history is not None:
        print("\n1. Training History Visualization")
        plot_training_history(history)
    
    
    print("\n2. Reconstruction Quality Examples")
    plot_reconstructions(autoencoder, X_test, class_names, labels_test, num_images=8)
    
    
    print("\n3. Latent Space Visualization")
    encoded_imgs = plot_latent_space_visualization(encoder, X_test, labels_test, class_names)
    
    
    print("\n4. Reconstruction Error Analysis")
    plot_reconstruction_error_heatmap(autoencoder, X_test, labels_test, class_names)
    
    
    print("\n5. Feature Maps Visualization (Sample Image)")
    sample_idx = np.random.choice(len(X_test))
    plot_feature_maps(encoder, X_test[sample_idx])
    
    
    print("\n6. Quantitative Evaluation Metrics")
    
    
    test_reconstructed = autoencoder.predict(X_test)
    train_reconstructed = autoencoder.predict(X_train)
    
    test_mse = np.mean((X_test - test_reconstructed) ** 2)
    train_mse = np.mean((X_train - train_reconstructed) ** 2)
    
    test_mae = np.mean(np.abs(X_test - test_reconstructed))
    train_mae = np.mean(np.abs(X_train - train_reconstructed))
    
    print(f"Training MSE: {train_mse:.6f}")
    print(f"Test MSE: {test_mse:.6f}")
    print(f"Training MAE: {train_mae:.6f}")
    print(f"Test MAE: {test_mae:.6f}")
    
    
    original_size = np.prod(X_test.shape[1:])
    compressed_size = encoder.output_shape[-1]  
    compression_ratio = original_size / compressed_size
    
    print(f"\nCompression Analysis:")
    print(f"Original image size: {original_size} pixels")
    print(f"Compressed representation: {compressed_size} dimensions")
    print(f"Compression ratio: {compression_ratio:.2f}:1")
    
    return {
        'test_mse': test_mse,
        'train_mse': train_mse,
        'test_mae': test_mae,
        'train_mae': train_mae,
        'compression_ratio': compression_ratio,
        'encoded_representations': encoded_imgs
    }


if __name__ == "__main__":
    
    dataset_path = "/kaggle/input/chessman-image-dataset/"  
    
    
    images, labels, class_names = load_chess_dataset(dataset_path, img_size=(128, 128))
    
    
    print("\n" + "="*50)
    print("DATASET INFORMATION")
    print("="*50)
    print(f"Dataset shape: {images.shape}")
    print(f"Image dimensions: {images.shape[1]}x{images.shape[2]}x{images.shape[3]}")
    print(f"Total number of images: {images.shape[0]}")
    print(f"Number of classes: {len(class_names)}")
    print(f"Class names: {class_names}")
    
    
    unique, counts = np.unique(labels, return_counts=True)
    print(f"\nClass distribution:")
    for i, (class_idx, count) in enumerate(zip(unique, counts)):
        print(f"  {class_names[class_idx]}: {count} images")
    
    
    X_train, X_test, labels_train, labels_test = train_test_split(images, labels, test_size=0.2, random_state=42, stratify=labels)
    
    print(f"\nTraining set size: {X_train.shape[0]}")
    print(f"Test set size: {X_test.shape[0]}")
    print(f"Training labels size: {labels_train.shape[0]}")
    print(f"Test labels size: {labels_test.shape[0]}")
    
    
    input_shape = images.shape[1:]  
    autoencoder, encoder, decoder = create_convolutional_autoencoder(input_shape, latent_dim=128)
    
    
    print("\n" + "="*50)
    print("AUTOENCODER ARCHITECTURE")
    print("="*50)
    print("\nFull Autoencoder Summary:")
    autoencoder.summary()
    
    print("\n" + "-"*30)
    print("Encoder Summary:")
    encoder.summary()
    
    print("\n" + "-"*30)
    print("Decoder Summary:")
    decoder.summary()
    
    
    autoencoder.compile(optimizer='adam', loss='mse', metrics=['mae'])
    
    
    print("\nDisplaying sample images...")
    plot_sample_images(images, labels, class_names)
    
    print("\nModel is ready for training!")
    print("To train the model, you can use:")
    print("history = autoencoder.fit(X_train, X_train, epochs=50, batch_size=32, validation_data=(X_test, X_test))")
    
    
    print("\nTraining the model...")
    history = autoencoder.fit(
        X_train, X_train,
        epochs=20,
        batch_size=32,
        validation_data=(X_test, X_test),
        verbose=1
    )
    
    
    evaluation_results = comprehensive_evaluation(
        autoencoder, encoder, decoder, 
        X_train, X_test, 
        labels_train, labels_test,  
        class_names, history
    )
    
    print("\nEvaluation completed!")
    
```
